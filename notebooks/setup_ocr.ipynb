```python
import cv2
import pytesseract
import matplotlib.pyplot as plt

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Load gambar dari folder dataset_dummy
image_path = '../dataset_dummy/struk 3.jpg' 
img = cv2.imread(image_path)

# Pre-processing: ubah ke grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Gaussian blur untuk menghilangkan bintik hitam 
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Pre-processing: adaptive thresholding 
thresh = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 21, 10)

# Tampilkan perbandingan gambar
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axes[0].set_title('Foto Struk Asli')
axes[1].imshow(thresh, cmap='gray')
axes[1].set_title('Hasil Filter OpenCV')
plt.show()

# Ekstraksi teks dengan tesseract
print("--- HASIL BACAAN MESIN OCR ---")
text = pytesseract.image_to_string(thresh, config='--psm 6')
print(text)
```
